```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
data = pd.read_csv("happiness.csv")
```

```python
data.head()
```

```python
data.describe
```

```python
features = data[['Happiness Score']]
```

```python
features
```

```python
label = data[['Happiness Rank']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, label, test_size = 0.2, random_state = 546)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
predictions
```

```python
model.predict([[7.4]])
```

```python
r2_score(y_test, predictions)
```
